visualize data

```python
fir = plt.figure()
fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(2,2, figsize=(8, 8))

ax1.plot(df["Length"], df["Rings"])
ax1.set_xlabel('Lenght')
ax1.set_ylabel('Diameter')

ax2.plot(df["Height"], df["Rings"])
ax2.set_xlabel('Height')
ax2.set_ylabel('Whole weight')

ax3.plot(df["sw"], df["Rings"])
ax3.set_xlabel('Shucked weight ')
ax3.set_ylabel('Viscera weight')

ax4.plot(df["sw.1"],  df["Rings"], 'b') 
ax4.set_xlabel('Shell weigh')
ax4.set_ylabel('Rings')
 

plt.show()
```

start trainig data

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import SGDClassifier

y = df["Rings"]
X = df.drop(['Sex', 'Rings'], axis=1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

clf = SGDClassifier(loss="hinge", penalty="l2", max_iter=5)
clf.fit(X_train, y_train)

print(clf.score(X_test, y_test))
```
